```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Dataset.csv', skiprows=4)

# Extract 2022 population data
pop_2022 = df[['Country Name', '2022']].dropna()

# Sort and get top 10
top10 = pop_2022.sort_values(by='2022', ascending=False).head(10)

# Format population in millions
top10['Population (Millions)'] = top10['2022'] / 1_000_000

# Plot - Horizontal Bar Chart
plt.figure(figsize=(12, 7))
bars = plt.barh(top10['Country Name'], top10['Population (Millions)'], color='mediumseagreen')
plt.title('Top 10 Most Populous Countries (2022)', fontsize=16)
plt.xlabel('Population (in Millions)')
plt.gca().invert_yaxis()  # Highest population at top
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Add value labels
for bar in bars:
    plt.text(bar.get_width() + 10, bar.get_y() + bar.get_height()/2,
             f'{bar.get_width():,.1f}', va='center')

# Save the figure
plt.tight_layout()
plt.savefig('top10_population_2022.png')
plt.show()
```
